```python
import pandas as pd
```

```python
df_syn_philly_findiff = pd.read_csv('syn_philly_findiff.csv')
df_helios_processed = pd.read_csv('df_helios_processed.csv')
df_syn_bw_findiff = pd.read_csv('syn_bw_findiff.csv')
df_syn_helios_findiff = pd.read_csv('syn_helios_findiff.csv')
df_bw_processed = pd.read_csv('df_bw_processed.csv')
df_syn_theta_findiff_3000_bs = pd.read_csv('syn_theta_findiff_3000_bs.csv')
df_theta_processed = pd.read_csv('df_theta_processed.csv')
df_philly_processed = pd.read_csv('df_philly_processed.csv')
```

```python
df_syn_bw_findiff.columns
```

```python

df_syn_philly_findiff.columns = ['u_id', 'user', 'gpu_num', 'cpu_num', 'node_num', 'interval', 'run_time', 'wall_time']
df_syn_bw_findiff.columns = ['u_id', 'user', 'gpu_num', 'cpu_num', 'node_num', 'interval', 'run_time', 'wall_time']
df_syn_helios_findiff.columns = ['u_id', 'user', 'gpu_num', 'cpu_num', 'node_num', 'interval', 'run_time', 'wall_time']
df_syn_theta_findiff_3000_bs.columns = ['u_id', 'user', 'gpu_num', 'cpu_num', 'node_num', 'interval', 'run_time', 'wall_time']
```

```python
df_syn_philly_findiff['u_id'] = df_syn_philly_findiff['u_id'].astype(int)
df_syn_philly_findiff['user'] = df_syn_philly_findiff['user'].astype(int)
df_syn_bw_findiff['u_id'] = df_syn_bw_findiff['u_id'].astype(int)
df_syn_bw_findiff['user'] = df_syn_bw_findiff['user'].astype(int)
df_syn_helios_findiff['u_id'] = df_syn_helios_findiff['u_id'].astype(int)
df_syn_helios_findiff['user'] = df_syn_helios_findiff['user'].astype(int)
df_syn_theta_findiff_3000_bs['u_id'] = df_syn_theta_findiff_3000_bs['u_id'].astype(int)
df_syn_theta_findiff_3000_bs['user'] = df_syn_theta_findiff_3000_bs['user'].astype(int)
```

```python
df_syn_philly_findiff = df_syn_philly_findiff.drop(columns=['wall_time', 'cpu_num'])
df_philly_processed = df_philly_processed.drop(columns=['wall_time', 'cpu_num'])
```

```python
df_helios_processed = df_helios_processed.drop(columns=['wall_time'])
df_syn_helios_findiff = df_syn_helios_findiff.drop(columns=['wall_time'])
```

```python
df_theta_processed = df_theta_processed.drop(columns=['gpu_num'])
df_syn_theta_findiff_3000_bs = df_syn_theta_findiff_3000_bs.drop(columns=['gpu_num'])
```

# BW Heatmaps

```python

import seaborn as sns
import matplotlib.pyplot as plt
combined_bw_df = pd.concat([df_bw_processed, df_syn_bw_findiff], ignore_index=True)
corr_matrix_bw = combined_bw_df.corr()
fig, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(corr_matrix_bw, annot=True, fmt=".2f", ax=ax, cmap="coolwarm")
plt.title("Heatmap comparing bw datasets")
plt.show()
```
