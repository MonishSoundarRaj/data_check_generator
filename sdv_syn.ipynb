```python
import pandas as pd
import numpy as np
```

```python
theta_df = pd.read_csv("data_theta.csv")
theta_df.drop("Unnamed: 0", axis=1, inplace=True)
theta_df.head()
```

```python
from sdv.metadata import SingleTableMetadata
```

```python
metadata = SingleTableMetadata()
```

```python
metadata.detect_from_dataframe(data=theta_df)
```

```python
print(metadata)
```

```python
theta_df.head()
```

```python
from sdv.single_table import CTGANSynthesizer

synthesizer = CTGANSynthesizer(
    metadata,
    enforce_rounding=False,
    epochs=30,
    verbose=True
)
synthesizer.fit(theta_df.head(14000))

synthetic_data = synthesizer.sample(num_rows=10000)
```

```python
print(synthetic_data)
```

```python
from google.colab import files
synthetic_data.to_csv("syn_ctgan_theta.csv")
files.download("syn_ctgan_theta.csv")
```

```python
from sdv.single_table import TVAESynthesizer

synthesizer_TVAE = TVAESynthesizer(
    metadata,
    enforce_min_max_values=True,
    enforce_rounding=False,
    epochs=100,
)

synthesizer_TVAE.fit(theta_df.head(13000))
synthetic_data_tvae = synthesizer_TVAE.sample(num_rows=1000)
```

```python
print(synthetic_data_tvae)
```

```python
from google.colab import files
synthetic_data_tvae.to_csv("syn_tvae_theta_epochs100.csv")
files.download("syn_tvae_theta_epochs100.csv")
```

```python
theta_df.columns
```

```python
from sdv.single_table import GaussianCopulaSynthesizer

synthesizer_gcs = GaussianCopulaSynthesizer(
    metadata, # required
    enforce_min_max_values=True,
    enforce_rounding=False,
    numerical_distributions={
        'user': 'gaussian_kde',
        'project': 'gaussian_kde',
    },
    default_distribution='beta'
)

synthesizer_gcs.fit(theta_df.head(13000))

synthetic_data_gcs = synthesizer_gcs.sample(num_rows=10000)
```

```python
synthetic_data_gcs.to_csv("data_theta_gcs.csv")
from google.colab import files
files.download("data_theta_gcs.csv")
```

```python
from sdv.single_table import CopulaGANSynthesizer

synthesizer_cgs = CopulaGANSynthesizer(
    metadata, # required
    enforce_min_max_values=True,
    enforce_rounding=False,
    numerical_distributions={
        'user': 'gaussian_kde',
        'project': 'gaussian_kde',
    },
    epochs=100,
    verbose=True
)

synthesizer_cgs.fit(theta_df.head(13000))

synthetic_data_cgs = synthesizer_cgs.sample(num_rows=10000)
```
