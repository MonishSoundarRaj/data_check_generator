```python
import pandas as pd
```

```python
df_bw = pd.read_csv("data_blue_waters.csv")
df_ph = pd.read_csv("data_philly.csv")
df_hl = pd.read_csv("data_helios.csv")
df_th = pd.read_csv("data_theta.csv")
df_syn_th = pd.read_csv("theta_data_training.csv")
```

```python
df_th = df_th.sort_values(by=['submit_time'])
df_th.reset_index(inplace=True, drop=True)
```

```python
df_ph = df_ph.sort_values(by=['submit_time'])
df_ph.reset_index(inplace=True, drop=True)
```

```python
df_hl = df_hl.sort_values(by=['submit_time'])
df_hl.reset_index(inplace=True, drop=True)
```

```python
df_bw = df_bw.sort_values(by=['submit_time'])
df_bw.reset_index(inplace=True, drop=True)
```

```python
df_th.head()
```

```python
first_submit_time = {
    "blue_waters": df_bw["submit_time"].min(),
    "philly": df_ph["submit_time"].min(),
    "helios": df_hl["submit_time"].min(),
    "theta": df_th["submit_time"].min(),
}

for platform, submit_time in first_submit_time.items():
    print(f"{platform}: {submit_time}")
```

```python
df_th.head()
```

```python
# Assuming the first job submit_time is given as 1539300 (Unix timestamp)
first_submit_time = 1539154542

# Since the interval is the difference between subsequent times,
# we should ensure that the first interval is 0 because it has no previous job.
df_syn_th.loc[0, 'interval'] = 0

# Convert intervals back to submit times, cumsum() function is used to create a cumulative sum of interval
df_syn_th['submit_time'] = df_syn_th['interval'].cumsum() + first_submit_time
```

```python
df_syn_th['submit_time'] = df_syn_th['submit_time'].astype(int)
df_syn_th.head()
# see now the submit_time in the unprocessed data and processed data match
```
