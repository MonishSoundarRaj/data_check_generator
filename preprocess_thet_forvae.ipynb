```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("data_theta.csv")
data.head()
```

```python
# prompt: drop Unnamed: 0 from the df

data.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Identify categorical and numerical columns
categorical_cols = ['state', 'new_status']
numerical_cols = ['gpu_num', 'cpu_num', 'node_num', 'submit_time', 'wait_time', 'run_time', 'wall_time', 'node_hour']

# Remove unnecessary columns
data = data.drop(columns=['job', 'user', 'project'])

# Define the pipeline for transformation
pipeline = ColumnTransformer([
    ('num', MinMaxScaler(), numerical_cols),
    ('cat', OneHotEncoder(), categorical_cols)
])

# Apply the transformations
data_prepared = pipeline.fit_transform(data)

# Display the shape of the transformed data
data_prepared.shape
```
