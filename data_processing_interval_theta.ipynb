```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data_theta_cgs.csv")
df.head(10)
```

```python
# prompt: how to ascend the rows based on the values of the intervals

df['interval'] = df['interval'].fillna(0)
df.sort_values('interval', inplace=True)
df.head(10)
```

```python
df = df.drop(columns=['Unnamed: 0'])
```

```python
# prompt: count the unique values in the interval column

df['interval'].unique().size
```

```python
# prompt: change the index of df as I have rearranged the data

df = df.reset_index(drop=True)
```

```python
df.head()
```

```python

initial_timestamp = 1640214735
df.loc[0, 'submit_time'] = initial_timestamp

df['submit_time'] = initial_timestamp + df['interval'].cumsum()

df.head()
```

```python
df = df.drop(columns=['interval'])
```
