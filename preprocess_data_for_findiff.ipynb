```python
import pandas as pd

df_bw = pd.read_csv("data_blue_waters.csv")
df_helios = pd.read_csv("data_helios.csv")
df_philly = pd.read_csv("data_philly.csv")
df_theta = pd.read_csv("data_theta.csv")
```

```python
# prompt: shape of all the dfs

print("Shape of df_bw:", df_bw.shape)
print("Shape of df_helios:", df_helios.shape)
print("Shape of df_philly:", df_philly.shape)
print("Shape of df_theta:", df_theta.shape)
```

```python
# prompt: print all the column names of the dfs separelty

print("Columns of df_bw:", df_bw.columns.to_list())
print("Columns of df_helios:", df_helios.columns.to_list())
print("Columns of df_philly:", df_philly.columns.to_list())
print("Columns of df_theta:", df_theta.columns.to_list())
```

```python
# prompt: now arrange the rows in dfs according to submit_time. and reset the indexs,

df_bw = df_bw.sort_values(by="submit_time").reset_index(drop=True)
df_helios = df_helios.sort_values(by="submit_time").reset_index(drop=True)
df_philly = df_philly.sort_values(by="submit_time").reset_index(drop=True)
df_theta = df_theta.sort_values(by="submit_time").reset_index(drop=True)
```

```python
df_theta.head()
```

```python
# prompt: add a Unnamed: 0.1 to df_theta, it should start from 0 and go upto the no of rows

df_theta["Unnamed: 0.1"] = range(len(df_theta))
df_theta.head()
```

```python
# prompt: drop Unnamed: 0

df_theta = df_theta.drop("Unnamed: 0", axis=1)
```

```python
df_theta.head()
```

```python
# prompt: change the name of the column "Unnamed: 0.1" to "Unnamed: 0" in df_theta

df_theta.rename(columns={"Unnamed: 0.1": "Unnamed: 0"}, inplace=True)
```

```python
df_theta.head()
```

```python
# prompt: in all the dfs only keep "user,  gpu_num, cpu_num, node_num, submit_time, run_time, wall_time, Unnamed: 0"

df_bw = df_bw[["Unnamed: 0", "user", "gpu_num", "cpu_num", "node_num", "submit_time", "run_time", "wall_time"]]
df_helios = df_helios[["Unnamed: 0", "user", "gpu_num", "cpu_num", "node_num", "submit_time", "run_time", "wall_time"]]
df_philly = df_philly[["Unnamed: 0", "user", "gpu_num", "cpu_num", "node_num", "submit_time", "run_time", "wall_time"]]
df_theta = df_theta[["Unnamed: 0", "user", "gpu_num", "cpu_num", "node_num", "submit_time", "run_time", "wall_time"]]

print("Shape of df_bw:", df_bw.shape)
print("Shape of df_helios:", df_helios.shape)
print("Shape of df_philly:", df_philly.shape)
print("Shape of df_theta:", df_theta.shape)
```

```python
# prompt: change the column name to Unnamed: 0 to u_id

df_bw.rename(columns={"Unnamed: 0": "u_id"}, inplace=True)
df_helios.rename(columns={"Unnamed: 0": "u_id"}, inplace=True)
df_philly.rename(columns={"Unnamed: 0": "u_id"}, inplace=True)
df_theta.rename(columns={"Unnamed: 0": "u_id"}, inplace=True)

df_theta.head()
```

```python
print("Columns of df_bw:", df_bw.columns.to_list())
print("Columns of df_helios:", df_helios.columns.to_list())
print("Columns of df_philly:", df_philly.columns.to_list())
print("Columns of df_theta:", df_theta.columns.to_list())
```

```python
# Initialize the dictionary for storing unique values and their mappings for 'user' column only
print(df_bw['user'].nunique())

unique_values = {'user': {}}
unique_values['user']['values'] = df_bw['user'].unique()
unique_values['user']['mapping'] = dict(zip(unique_values['user']['values'], range(1, len(unique_values['user']['values']) + 1)))

# Update the 'user' column in the DataFrame based on the mapping
for i, row in df_bw.iterrows():
    df_bw.at[i, 'user'] = unique_values['user']['mapping'][row['user']]

# Convert the 'user' column to integer type
df_bw['user'] = df_bw['user'].astype(int)

# Print the maximum value of the 'user' column
print(df_bw['user'].max())
```

```python
# Initialize the dictionary for storing unique values and their mappings for 'user' column only
print(df_theta['user'].nunique())
unique_values = {'user': {}}
unique_values['user']['values'] = df_theta['user'].unique()
unique_values['user']['mapping'] = dict(zip(unique_values['user']['values'], range(1, len(unique_values['user']['values']) + 1)))

# Update the 'user' column in the DataFrame based on the mapping
for i, row in df_theta.iterrows():
    df_theta.at[i, 'user'] = unique_values['user']['mapping'][row['user']]

# Convert the 'user' column to integer type
df_theta['user'] = df_theta['user'].astype(int)

# Print the maximum value of the 'user' column
print(df_theta['user'].max())
```

```python
# Initialize the dictionary for storing unique values and their mappings for 'user' column only
print(df_helios['user'].nunique())
unique_values = {'user': {}}
unique_values['user']['values'] = df_helios['user'].unique()
unique_values['user']['mapping'] = dict(zip(unique_values['user']['values'], range(1, len(unique_values['user']['values']) + 1)))

# Update the 'user' column in the DataFrame based on the mapping
for i, row in df_helios.iterrows():
    df_helios.at[i, 'user'] = unique_values['user']['mapping'][row['user']]

# Convert the 'user' column to integer type
df_helios['user'] = df_helios['user'].astype(int)

# Print the maximum value of the 'user' column
print(df_helios['user'].max())
```

```python
# Initialize the dictionary for storing unique values and their mappings for 'user' column only
print(df_philly['user'].nunique())
unique_values = {'user': {}}
unique_values['user']['values'] = df_philly['user'].unique()
unique_values['user']['mapping'] = dict(zip(unique_values['user']['values'], range(1, len(unique_values['user']['values']) + 1)))

# Update the 'user' column in the DataFrame based on the mapping
for i, row in df_philly.iterrows():
    df_philly.at[i, 'user'] = unique_values['user']['mapping'][row['user']]

# Convert the 'user' column to integer type
df_philly['user'] = df_philly['user'].astype(int)

# Print the maximum value of the 'user' column
print(df_philly['user'].max())
```

```python
# Convert submit_time to interval

df_bw['interval'] = df_bw['submit_time'] - df_bw['submit_time'].shift(1)
df_bw['interval'] = df_bw['interval'].fillna(0)
```

```python
# Convert submit_time to interval

df_theta['interval'] = df_theta['submit_time'] - df_theta['submit_time'].shift(1)
df_theta['interval'] = df_theta['interval'].fillna(0)
```

```python
# Convert submit_time to interval

df_helios['interval'] = df_helios['submit_time'] - df_helios['submit_time'].shift(1)
df_helios['interval'] = df_helios['interval'].fillna(0)
```

```python
# Convert submit_time to interval

df_philly['interval'] = df_philly['submit_time'] - df_philly['submit_time'].shift(1)
df_philly['interval'] = df_philly['interval'].fillna(0)
```

```python
df_bw = df_bw[["u_id", "user", "gpu_num", "cpu_num", "node_num", "interval", "run_time", "wall_time"]]
df_helios = df_helios[["u_id", "user", "gpu_num", "cpu_num", "node_num", "interval", "run_time", "wall_time"]]
df_philly = df_philly[["u_id", "user", "gpu_num", "cpu_num", "node_num", "interval", "run_time", "wall_time"]]
df_theta = df_theta[["u_id", "user", "gpu_num", "cpu_num", "node_num", "interval", "run_time", "wall_time"]]

print("Shape of df_bw:", df_bw.shape)
print("Shape of df_helios:", df_helios.shape)
print("Shape of df_philly:", df_philly.shape)
print("Shape of df_theta:", df_theta.shape)
```

```python
# prompt: convert all dfs to csvs

df_bw.to_csv("df_bw_processed.csv", index=False)
df_helios.to_csv("df_helios_processed.csv", index=False)
df_philly.to_csv("df_philly_processed.csv", index=False)
df_theta.to_csv("df_theta_processed.csv", index=False)
```
