```python
import pandas as pd
import numpy as np

df = pd.read_csv("data_blue_waters.csv")
p_df = pd.read_csv("data_philly.csv")
```

```python
df.head(10)
```

```python
p_df.head(10)
```

```python
# prompt: remove wait_time and Unnamed: 0 from df and p_df

df = df.drop(['wait_time', 'Unnamed: 0'], axis=1)
p_df = p_df.drop(['wait_time', 'Unnamed: 0'], axis=1)
```

```python
# prompt: arrange the data in ascending order of submit_time of df and p_df

df = df.sort_values(by='submit_time', ascending=True)
p_df = p_df.sort_values(by='submit_time', ascending=True)
```

```python
p_df.head()
```

```python
# prompt: okay, now apply this logic a-a.shift(1) for submit_time

df['interval'] = df['submit_time'] - df['submit_time'].shift(1)
p_df['interval'] = p_df['submit_time'] - p_df['submit_time'].shift(1)
```

```python
df.head()
```

```python
p_df.head()
```

```python
# prompt: check if there are any negative number in submit_time of df and P_df

negative_submit_times_df = df['interval'] < 0
negative_submit_times_p_df = p_df['interval'] < 0

print(f"Number of negative submit times in df: {negative_submit_times_df.sum()}")
print(f"Number of negative submit times in p_df: {negative_submit_times_p_df.sum()}")
```

```python
# prompt: drop submit_time from df and p_df

df = df.drop('submit_time', axis=1)
p_df = p_df.drop('submit_time', axis=1)
```

```python
# prompt: check now many NaN are there in internval column

df['interval'].isna().sum()
```

```python
# prompt: change the NaN with 0 in interval column of df and p_df

df['interval'] = df['interval'].fillna(0)
p_df['interval'] = p_df['interval'].fillna(0)
```

```python
df.to_csv("training_data_bw.csv")
p_df.to_csv("training_data_philly.csv")
```
